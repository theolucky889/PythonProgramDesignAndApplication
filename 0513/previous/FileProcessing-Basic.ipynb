```python
'''
檔案的操作:
讀寫檔案。
讀檔: 表示該檔案已經存在。
寫檔: 表示要創建一個新檔，但也會把已經存在的同名檔案覆蓋掉。請注意使用時要先做好備份！
read: 'r'  for read.
wirte: 'w'  for write
append: 'a'  for append

#-------------------------------------------------------------------------------
step 1: 開啟檔案
  fp = open(file_name, mode[, encoding='utf-8'])  # utf-8 的編碼格式。
  mode: r, w, a, ...
  其中的 fp 只是一個承接回傳內容的變數名稱，可以自訂，例如僅寫 f，或以其他名稱替代亦可。
  
step 2: 進行讀寫操作
  fp.read(), fp.readline(), fp.readlines()
  fp.write()
  
step 3: 關閉檔案
  fp.close()  <--如果未正常關閉開啟的檔案，有可能導致檔案不全，稍後無法開啟。
#-------------------------------------------------------------------------------  
'''
# 示範使用 write 將資料寫入檔案: write 什麼，就存什麼。

# >>> step 1:
fname = "output.txt"
fp = open(fname, "w")  # 有時需要明確地指定編碼的形式: big5 or utf-8。

# >>> step 2:
fp.write("這是一個文字檔案！")

# >>> step 3:
fp.close()  # 記得要關閉檔案，以完整地儲存內容！

print("Done!")

# dir(fp)
```

```python
'''
將下列資料寫入檔案: 方式 1 
1
12
123
1234
12345
'''

# >>> step 1:
fname = "output.txt"
fp = open(fname, "w")

# >>> step 2:
numstr = ""
for k in range(1, 6):
    numstr += str(k)
    fp.write(numstr)  # 結果對嗎?  <--需要再加上換行符號 \n！
#     fp.write(numstr + '\n')
#     fp.write('\n' + numstr)  # 參考下個的程式碼。
    
# >>> step 3:
fp.close()

print("Done!")
```

```python
'''
將下列資料寫入檔案: 方式 2 
1
12
123
1234
12345
'''

# 使用變數將內容逐項合成起來，最後再將該變數的內容存入檔案中。

# step 1:
fname = "output.txt"
fp = open(fname, "w")

# step 2:
result = ""
numstr = ""
for k in range(1, 6):
    numstr += str(k)
    if result == "":  # 接著要加入第一列資料。
        result = numstr
    else:
        result += '\n' + numstr

# 最後再一次性寫入檔案:
fp.write(result)

# step 3:
fp.close()

# 剛剛的 result 仍然還存在要輸出到檔案的內容！
print(result)
```

```python
def table9x9():
    result = ""
    for r in range(1, 10):
        for c in range(1, 10):
            if c == 1:
                result += f"{r}x{c}={r*c} "
            else:
                result += f"{r}x{c}={r*c:2} "
        result += "\n"
    return result;

print( table9x9() )

'''
請將九九乘法表的內容，存入檔案 Table9x9.txt 中。
'''
```

```python
'''
因為關閉檔案這個程序很重要，但卻是最後要做的部分，因此經常會被遺忘！
所以Python提供一個寫作方式: with open(...) as，
讓我們只要專注在檔案的開啟(包含讀與寫)與操作，且在結束with後會自動幫我們關閉檔案。

with open(file_name, mode, encoding="coding-mode") as fp:
    # do your file processing here
    
When leaving the with block, python closes the file for us automatically.   
'''

fname = "54007.dat"
with open(fname, "w") as fp:
    fp.write("王大同")
    fp.write("\n54007")
    fp.write("\n不分系")

#---------------------------------------------
# fname = "54007-utf-8.dat"
# with open(fname, "w", encoding='utf-8') as fp:
#     fp.write("王大同")
#     fp.write("\n54007")
#     fp.write("\n不分系")
    
print("Done!")
```
