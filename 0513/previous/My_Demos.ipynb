**星座的判別**

[星座的英文](https://www.vogue.com.tw/lifestyle/article/12%E6%98%9F%E5%BA%A7-%E8%8B%B1%E6%96%87)

類似的單字還有：star sign「星座」、zodiac sign「星座」、horoscope「星相運勢」

````python
'''
星座:日期範圍
魔羯座:12.22~1.19 
水瓶座:1.20~2.18  
雙魚座:2.19~3.20  
牡羊座:3.21~4.19  
金牛座:4.20~5.20  
雙子座:5.21~6.20  
巨蟹座:6.21~7.22  
獅子座:7.23~8.22  
處女座:8.23~9.22  
天秤座:9.23~10.22 
天蠍座:10.23~11.21
射手座:11.22~12.21

請參考上述的星座資料，寫一個Python函式，讓使用者輸入月與日的資料後，
接著會顯示這一天對應的星座名稱，同時印出該星座的是日期範圍，方便讓使用者再次確認。
'''

msg = """
星座:日期範圍
魔羯座:12.22~1.19 
水瓶座:1.20~2.18  
雙魚座:2.19~3.20  
牡羊座:3.21~4.19  
金牛座:4.20~5.20  
雙子座:5.21~6.20  
巨蟹座:6.21~7.22  
獅子座:7.23~8.22  
處女座:8.23~9.22  
天秤座:9.23~10.22 
天蠍座:10.23~11.21
射手座:11.22~12.21
"""
# 先將各個星座的資料取出:
lines = msg.split()
show('lines')  # => '魔羯座:12.22~1.19'
markline()

items = [line.split(':') for line in lines]  # 將星座的部分拆解出來，索引 1 表示一月，以空間換取時間，。
show('items')  # => ['魔羯座', '12.22~1.19']
markline()

zodiac_sign = [it[0] for it in items]  # 取出星座的名稱
show('zodiac_sign')

zodiac_sign.append('魔羯座')  # 把'魔羯座'，附加到12月的'射手座'後面，做為'射手座'的下一個星座。
show('zodiac_sign')
markline()

# 取出記錄星座截止的日期，超過該日期，就變成'下一個'星座！
print("取出星座的截止日期:")
zodiac_to = [it[1].split('.')[-1] for it in items]  # [-1]: 表示最後一個元素。
show('zodiac_to')

# 將字串轉換成數字:
print("將第一項之後的內容，轉換成數字:")
zodiac_to = [int(zt) for zt in zodiac_to[1:]]  # 排除 '日期範圍' 這一項。
show('zodiac_to')

# 為截止日期插入無關緊要的第一項，使得索引跟月份一致。
print("插入第一項，讓索引跟月份一致:")
zodiac_to.insert(0, 0)  # 插入一個開頭數字，讓'索引值'與'月份'相同。
show('zodiac_to')
markline()
````

````python
# 先顯示目前的整合結果:
show('zodiac_sign')
show('zodiac_to')
markline()

inp = input("輸入 月/日: ");
inp = inp.split('/')

month, date = int(inp[0]), int(inp[1])

# 由月份來決定星座名稱，但如果 date 大於 輸入的星座日期 zodiac_to[month]，則會變成下一個星座。 
if date <= zodiac_to[month]:
    zodiac = zodiac_sign[month]
else:
    zodiac = zodiac_sign[month + 1]  # next
    
show('zodiac')
````

````python
'''
About dict
{ key1:value1, key2:value2, ....}
'''

# 宣告並給定 dict 的初值:
dct = {'a': 1, 'b': 2, 'c': 3}
print(dct)

# 預設只會輸出 key 的值:
for d in dct:
    print(d)
markline()

print('a' in dct)  # 只會比對 key 的值！
markline()

# dict 從無到有的設定: dict_name = {};  dict_name[key] = value, ...
# 或是說，可以可以為 dict 添加新的元素，即給新的 key:value 對。
about_me = {}

about_me['name'] = '王大同'
about_me['height'] = 170
about_me['weight'] = 70

print(about_me)
````

````python
'''
namestr = "廖靜宜、吳百凱、陳懿沛、蔡仁芸、王彥君、陳鈺瑋、林珊群、許萬昌、黃靜東、陳佩舜、林怡喜"
姓名資料如上，請Python程式將同姓的人名集合起來。<hr>
其中，我們假設
name = 姓氏 + 名字
     = name[0] + name[1:]
再逐一確認姓氏是否已經存在? 
1) 如果是，就將名字附加上去，名字間需要用逗點隔開。
2) 否則，另行將此姓氏新增至dict中，同時把名字當成是它的值，來做成記錄。
  
最後，請完全依照下列的格式，輸出完成拆解分析後的結果。
廖: 靜宜
吳: 百凱
陳: 懿沛,鈺瑋,佩舜
蔡: 仁芸
王: 彥君
林: 珊群,怡喜
許: 萬昌
黃: 靜東
'''
namestr = "廖靜宜、吳百凱、陳懿沛、蔡仁芸、王彥君、陳鈺瑋、林珊群、許萬昌、黃靜東、陳佩舜、林怡喜"

# 拆解來字的姓名資料:
names = namestr.split('、')  # 可以不必排序！
show('names')

surnames = {}  # 取姓名的第一個字當成姓氏，也就是 key。
for name in names:
    if name[0] in surnames:  # 姓氏如果已經存在 dict surnames 中，則將新的名字前置一個逗點，再附加在原先名字的後面。
        surnames[name[0]] += ',' + name[1:]  # 再原先的內容後，附加上新的名字。
    else:
        surnames[name[0]] = name[1:]  # 全新的姓名項。

show('surnames')
markline()

for key in surnames:
    print(f"{key}: {surnames[key]}")
````

````python
'''
3)
有時候在 import 一個 module 時，需要的僅是其中的一個函式庫或類別，這時為了更簡潔清楚地引用，
可以只將這些部份標記出來即可。
不過使用這種方式，我們雖然簡化了對這些函式的呼叫，但卻增加了與其他函式名稱重複，而導致錯誤的可能性！
'''
from datetime import datetime, timedelta  # 從同一 module 中抽出多個類別！
now = datetime.now()  # using datetime

# 透過設定部分參數的值，來進行引用。 如下的 weeks=1。
# 日期或時間差的應用:
one_week_later = timedelta(weeks=1)  # using datetimedelta with specified parameter

print(now - one_week_later)
print(now)
print(now + one_week_later)
````
