```python
'''
改成使用 while 來實作。
記得要改變條件，避免造成無窮迴圈的狀況！
'''
count = 0
total = 0  # 初始化資料！很重要！
print("請輸入5個整數:")
while count < 5:
    count += 1  # 修改條件很重要，不可忘記！
    inp = input(f"number {count}:")
    value = int(inp)
    total += value
    
print(f"總和為 {total}")   
```

```python
'''
無窮迴圈的應用
透過輸入資料的值，來決定是否結束。
請持續輸入資料，直到輸入的值是 0 為止！
'''
count = 0
total = 0  # 初始化資料！很重要！
print("請輸入整數來進行合計 或 輸入 0 結束程式執行:")
while True:    
    inp = input(f"number {count + 1}:")
    value = int(inp)
    if value != 0:
        count += 1
        total += value
    else:
        break
    
print(f"共輸入{count}個數字，其總和為 {total}")    
```

```python
'''
Q: 是不是可以使用單一變數名稱，來存多個資料呢?
A: 串列(list) in Python
(* 在其他程式語言中，這種型態的變數通常稱為陣列(array)，用來存多個相同型態的內容。)

Python中可以使用串列，且命名的規則跟一般變數相同。

Q: 如何取得其中儲存的不同內容呢?
A: 變數名稱[索引值]: 使用變數名稱附加索引，其值從 0 開始，依序增加 1。

取變數名稱時，我們通常會使用最接近的英文名，
例如，字串 -> str，但 str 是保留字，原則上不應該用！
不過，Python 卻允許使用！因此會造成稍後難以發覺的錯誤！
如果真的要用的話，可以在其後加上'底線與編號'，這樣某種程度採用了這個名稱，同時也避免了與原先的保留字衝突！
因此，字串不要用 str，而是儘可能改成 str_1, str_a。
'''
# 宣告一個串列:
lst_1 = [1, 2, 3]  # 命名為，也其內依序存著三個資料: 1, 2, 3。
show('lst_1')
newline()

# 取出其中的資料: 需邀透過索引來存取其中的資料，索引值從 0 開始，依序增加 1。
print(lst_1[0], lst_1[1], lst_1[2])

# 也可以改變串列內容的值:
lst_1[0] = lst_1[0] * 2         # lst_1[0] = 2
lst_1[1] = lst_1[1] + lst_1[2]  # lst_1[1] = 2+3 = 5
show('lst_1')

# 使用 len() 來取得串列共包含多少元素。
show('len(lst_1)')
```

```python
'''
range(start, cut, step)
預設值: 0           1   

從 start 開始，在 cut 處切斷(不包含的意思)，每次的增量值都是 step。
start <= start, start+step, start+2*step, ...,  < cut

range(N) = range(0:N) => 0, 1, 2, 3, ..., N-1
'''
for k in range(3):
    print("Hello!")
    
print()
for _ in range(3):  # 不需要用到 range 產生的數字時，可將變數以底線(_)表示。
    print("Hello!")
print()
```
