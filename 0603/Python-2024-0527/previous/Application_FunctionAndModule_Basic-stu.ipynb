```python
'''
星座的判斷規則的判斷:
我們可以有很多方式，來判斷某個日期是歸屬於星座。
用開始日期或結束日期均可。
重點在用那個日期來進行分段，來區分不同的星座。

以 魔羯座:12.22~1.19 為例來說明:
1) 採用開始日期 12.22
   如果給的月份是12月，那麼日期大於22日，就是魔羯座，否則會變成射手座(前一個)。
2) 採用結束日期 1.19
   如果給的月份是1月，那麼日期小於或等於19日，就是魔羯座，否則會變成水瓶座(後一個)。
'''
```

```python
'''
dict(dictionary): 字典結構。
{key_1:valye_1 [, key_2:value_2, ...]}
{key_1:value_1}
{key_1:value_1, key_2:value_2, ...}

上述的key也可以稱為屬性，其值可以是字串或數字。
儘管可以是數字，但並不會因數字而改變內容的先後順序。
'''
def bmi(height, weight):
    return weight / (height/100)**2


def show_list(lst):
    print(f"{lst[0]}身高{lst[1]}cm, 體重{lst[2]}kg，BMI= {bmi(lst[1], lst[2]):.1f}")
    
    
def show_dict(dct):
    print(f"{dct['name']}身高{dct['height']}cm, 體重{dct['weight']}kg，BMI= \
            {bmi(dct['height'], dct['weight']):.1f}")
    
# 要儲存某人的姓名、身高與體重
# case 1: 使用 串列(list)
someone = ['王大同', 170, 70]
show_list(someone)

# case 2: 使用 字典(dict)
# someone = {'name': '王大同', 'height': 170, 'weight':70 } 
# show_dict(someone)
```
