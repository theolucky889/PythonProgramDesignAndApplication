````python
# Problem 1
def fill_table(data, cols):
    """
    將一維串列資料依序填入 HTML 表格中。
    data: 一維串列資料。
    cols: 每列包含的元素個數。
    """
    rows = (len(data) + cols - 1) // cols
    
    html = "<style> th { padding: 6px; } </style>" + \
           "<table border=1>"
    
    for r in range(rows):
        html += "\n<tr>" 
        for c in range(cols):
            index = r * cols + c
            if index < len(data):
                html += f"<th>{data[index]}</th>"
            else:
                html += "<th></th>" 
        html += "</tr>"
    
    html += "\n</table>"
    
    return html

data = [k for k in range(1, 21)]
cols = 5
html_table = fill_table(data, cols)

html_file = 'table_output.html'
with open(html_file, 'w', encoding='utf-8') as file:
    file.write(html_table)

html_link = f'<a href="{html_file}">View HTML Table</a>'
print(html_link)


````

```
接著請找出我們學校中有日間部四技的系，將校名、系別與新生人數(含男女生)，填入 HTML 的表格中，部分輸出結果如下圖所示。
(可以使用 requests 或 pandas 來讀取資料檔案。)
```
![北科部分科系的新生人數](./images/TaipeiTech-Departments.jpg)


````python
import requests
import pandas as pd

url = "https://stats.moe.gov.tw/files/detail/112/112_students.csv"
response = requests.get(url)
response.encoding = 'utf-8'

with open("112_students.csv", "w", encoding='utf-8') as f:
    f.write(response.text)

df = pd.read_csv("112_students.csv")

school_dept = "國立臺北科技大學"
ntut_school_dept = df[
    (df["學校名稱"] == school_dept)
    & (df["日間∕進修別"] == "D 日")
    & (df["等級別"] == "B 四技")
]

html_table = """
<!DOCTYPE html>
<html>
<head>
<meta charset="UTF-8">
<title>112學年國立臺北科技大學日間部四技新生人數</title>
<style>
    table {{
        border-collapse: collapse;
        width: 100%;
    }}
    th, td {{
        border: 1px solid black;
        text-align: center;
        padding: 8px;
        white-space: nowrap;
    }}
    th {{
        background-color: #f2f2f2;
    }}
    tr:nth-child(even) {{
        background-color: #f9f9f9;
    }}
</style>
</head>
<body>

<h1>112學年國立臺北科技大學日間部四技新生人數</h1>
<table>
    <tr>
        <th>校名</th>
        <th>系別</th>
        <th>一年級男生</th>
        <th>一年級女生</th>
    </tr>
    {}
</table>
</body>
</html>
"""

table_rows = []
for index, row in ntut_school_dept.iterrows():
    table_rows.append(
        f"""
    <tr>
        <td>{row['學校名稱']}</td>
        <td>{row['科系名稱']}</td>
        <td>{row['一年級男生']}</td>
        <td>{row['一年級女生']}</td>
    </tr>
    """
    )

html_table = html_table.format("".join(table_rows))

with open("112_ntut_school_dept.html", "w", encoding='utf-8') as f:
    f.write(html_table)
````
